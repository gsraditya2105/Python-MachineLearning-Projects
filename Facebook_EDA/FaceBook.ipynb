##  Pre-Processing of Data

```python
fb_df.columns
```

```python
sns.distplot(fb_df.age)
```

```python
# - drop users with age >80 which is about 5.78% of information lost
# - more information is concentrated towards age range 15-80
drop_rec_per = fb_df[fb_df.age > 80].shape[0] / fb_df.shape[0] * 100
print(drop_rec_per)
fb_df.drop(fb_df[fb_df.age > 80].index, axis=0, inplace=True)
```

```python
# - combine dob_day, dob_month and dob_year  - convert into datetime object
fb_df['date_of_birth'] = fb_df['dob_day'].astype(str) + str("/") + fb_df[
    'dob_month'].astype(str) + str("/") + fb_df['dob_year'].astype(str)
fb_df['date_of_birth'] = pd.to_datetime(fb_df['date_of_birth'],
                                        format='%d/%m/%Y')
```

```python
# drop columns userid, dob_day, dob_month, dob_year
fb_df.drop(['userid', 'dob_day', 'dob_month', 'dob_year'],
           axis=1,
           inplace=True)
```

```python
# - Find Missing Values in each column
fb_df.isna().sum()
```

```python
fb_df.gender.unique()
```

```python
# fill all missing values in gender column with 'male'
fb_df.gender.fillna('male', inplace=True)
```

```python
# drop the duplicates
fb_df.drop_duplicates(inplace=True, ignore_index=True)
```

```python
fb_df.isna().sum()
```

## Univariate Analysis

```python
# - adding age_group column
def setgroup(age):
    if (age > 10 and age <= 20):
        return "11-20"
    elif (age > 20 and age <= 30):
        return "21-30"
    elif (age > 30 and age <= 40):
        return "31-40"
    elif (age > 40 and age <= 50):
        return "41-50"
    elif (age > 50 and age <= 60):
        return "51-60"
    elif (age > 60 and age <= 70):
        return "61-70"
    elif (age > 70 and age <= 80):
        return "71-80"


fb_df['age_group'] = fb_df['age'].apply(setgroup)
```

## Bivariate Analysis

### which age group have more number of male and female users ?

```python
sns.countplot(x='age_group', data=fb_df, hue='gender', palette='rainbow')
plt.title("Gender categorical count per age group")
plt.savefig("fig-7.jpg")
```

### Who have more number of friends ? Male or Female ?

```python
f, ax = plt.subplots(1, 2, figsize=(12, 8))
fb_df.groupby('gender')['friend_count'].apply(sum).plot(
    kind='pie',
    explode=[0.1, 0.1],
    ax=ax[0],
    autopct='%.2f%%',
    legend=True,
    wedgeprops={'width': 0.15},
    shadow=True,
    cmap='rainbow')
fb_df.groupby('gender')['friendships_initiated'].apply(sum).plot(
    kind='pie',
    explode=[0.1, 0.1],
    ax=ax[1],
    autopct='%.2f%%',
    legend=True,
    wedgeprops={'width': 0.15},
    shadow=True,
    cmap='rainbow')
plt.savefig("fig-8.jpg")
```

### who have more likes ? Male or Female ? web or mobile ?

```python
f, ax = plt.subplots(1, 3, figsize=(16, 12))
fb_df.groupby('gender')['likes'].apply(sum).plot(kind='pie',
                                                 explode=[0.1, 0.1],
                                                 autopct='%.2f%%',
                                                 legend=True,
                                                 wedgeprops={'width': 0.15},
                                                 shadow=True,
                                                 cmap='viridis',
                                                 ax=ax[0])

m_www = (fb_df[fb_df.gender == "male"]['www_likes'].sum() /
         fb_df[fb_df.gender == "male"]['likes'].sum() * 100).round(2)
m_mobile = (fb_df[fb_df.gender == "male"]['mobile_likes'].sum() /
            fb_df[fb_df.gender == "male"]['likes'].sum() * 100).round(2)
f_www = (fb_df[fb_df.gender == "female"]['www_likes'].sum() /
         fb_df[fb_df.gender == "female"]['likes'].sum() * 100).round(2)
f_mobile = (fb_df[fb_df.gender == "female"]['mobile_likes'].sum() /
            fb_df[fb_df.gender == "female"]['likes'].sum() * 100).round(2)
df_new = pd.DataFrame({
    "male": [m_www, m_mobile],
    "female": [f_www, f_mobile]
},
    index=["www_likes", "mobile_likes"])

df_new.male.plot(kind='pie',
                 subplots=True,
                 explode=[0.1, 0.1],
                 autopct='%.2f%%',
                 legend=True,
                 wedgeprops={'width': 0.15},
                 shadow=True,
                 cmap='RdYlGn',
                 ax=ax[1])

df_new.female.plot(kind='pie',
                   subplots=True,
                   explode=[0.1, 0.1],
                   autopct='%.2f%%',
                   legend=True,
                   wedgeprops={'width': 0.15},
                   shadow=True,
                   cmap='RdYlGn',
                   ax=ax[2])

plt.savefig("fig-9.jpg")
```

### who have more likes_received ? Male or Female ? web or mobile ?

```python
f, ax = plt.subplots(1, 3, figsize=(16, 12))
fb_df.groupby('gender')['likes_received'].apply(sum).plot(
    kind='pie',
    explode=[0.1, 0.1],
    autopct='%.2f%%',
    legend=True,
    wedgeprops={'width': 0.15},
    shadow=True,
    cmap='viridis',
    ax=ax[0])

m_www = (fb_df[fb_df.gender == "male"]['www_likes_received'].sum() /
         fb_df[fb_df.gender == "male"]['likes_received'].sum() * 100).round(2)
m_mobile = (fb_df[fb_df.gender == "male"]['mobile_likes_received'].sum() /
            fb_df[fb_df.gender == "male"]['likes_received'].sum() *
            100).round(2)
f_www = (fb_df[fb_df.gender == "female"]['www_likes_received'].sum() /
         fb_df[fb_df.gender == "female"]['likes_received'].sum() *
         100).round(2)
f_mobile = (fb_df[fb_df.gender == "female"]['mobile_likes_received'].sum() /
            fb_df[fb_df.gender == "female"]['likes_received'].sum() *
            100).round(2)
df_new = pd.DataFrame({
    "male": [m_www, m_mobile],
    "female": [f_www, f_mobile]
},
    index=["www_likes", "mobile_likes"])

df_new.male.plot(kind='pie',
                 subplots=True,
                 explode=[0.1, 0.1],
                 autopct='%.2f%%',
                 legend=True,
                 wedgeprops={'width': 0.15},
                 shadow=True,
                 cmap='RdYlGn',
                 ax=ax[1])

df_new.female.plot(kind='pie',
                   subplots=True,
                   explode=[0.1, 0.1],
                   autopct='%.2f%%',
                   legend=True,
                   wedgeprops={'width': 0.15},
                   shadow=True,
                   cmap='RdYlGn',
                   ax=ax[2])

plt.savefig("fig-10.jpg")
```

### parallel_coordinates

```python
df_pc = fb_df.copy()
for colname in fb_df.columns:
    if colname in [
            'tenure', 'friend_count', 'friendships_initiated', 'likes',
            'likes_received', 'mobile_likes', 'mobile_likes_received',
            'www_likes', 'www_likes_received'
    ]:
        df_pc[colname] = (df_pc[colname] - df_pc[colname].min()) / (
            df_pc[colname].max() - df_pc[colname].min())
plt.figure(figsize=(12, 8))
parallel_coordinates(df_pc,
                     class_column='gender',
                     colormap='rainbow_r',
                     cols=[
                         'tenure', 'friend_count', 'friendships_initiated',
                         'likes', 'likes_received', 'mobile_likes',
                         'mobile_likes_received', 'www_likes',
                         'www_likes_received'
                     ])
plt.savefig("fig-16.jpg")
```
